```python
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', match='Postal Code')
len(table)
```

```python
df = table[0]
df.head(15)
```

```python
df.drop(df.index[df['Borough'] == 'Not assigned'], inplace = True)

# Reset Index
df = df.reset_index(drop=True)

df.head()
```

```python
df = df.groupby(['Postal Code', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()

df.head(50)
```

```python
df.loc[df['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = df['Borough']

df.head()
```

```python
df.to_csv(r'df_wiki.csv')
```

```python
url = "http://cocl.us/Geospatial_data"
df1 = pd.read_csv(url)

df1.head()
```

```python
df2 = pd.merge(df, df1, left_on = 'Postal Code', right_on = 'Postal Code', how = 'left')


df2.head()
```

```python
df2.to_csv(r'df_geo.csv')
```
